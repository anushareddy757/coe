```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
sns.get_dataset_names()
```

```python
df=sns.load_dataset("tips")
df
```

insights
1.In the above datset there are total 244 rows and 7 columns with no missing values
2.The total bill amounts vary from low to high values, while tips also show a reasonable range, indicating diverse customer spending.
3.The restaurant receives more customers on weekends compared to weekdays.
4.Dinner hours are more popular than lunch hours among customers.
5.Small groups are the most common, showing people usually dine in pairs or trios.
6.Customers with higher bills tend to give higher tips, though not always proportionally.
